### Lists

```python
# Lists
# Definition: List are the collection of objects.
list1=['Red','Green']
list2=[10,20,30]
# Operators
# Summary of List Operations:
'''
Multiplication operator *:
- The multiplication operator is used to repeat a list n times.
- Syntax:
    - list*n
'''
print("Multiplication operator example")
print(list1*2)
'''
Concatenation operator +:
- The concatenation operator is used to add elements of two lists.
- Syntax:
    - list1+list2
'''
print("Concatenation operator example")
print(list1+list2)
'''
Length operator:
- The length operator is used to find the length of the list.
- Syntax:
    - len(list1)
'''
print("Length operator example")
print(len(list1))
'''
Indexing:
- The indexing is to pick elements from the list.
- Syntax:
    - list[index}
'''
print("Indexing example")
print(list1[-1])
'''
Slicing:
- The slicing is used to move through the list.
- Syntax:
    - list[start:end:inc]
'''
print("Slicing example")
print(list2[::2])
'''
Min:
- The min function is used to find the minimum element in the list.
- Syntax:
    - min(list)
'''
print("Min function example")
print(min(list2))
'''
Max:
- The max function is used to find the maximum element in the list.
- Syntax:
    - max(list)
'''
print("Min function example")
print(max(list2))
'''
Sum:
- The sum finction is used to find the sum of the elements in the list.
- Syntax:
    - sum(list)
'''
print("Sum function example")
print(sum(list2))
'''
Membership operator in:
- The in operator is used to find whether the element in the list or nor.
- Syntax:
    - obj in list
'''
print("Membership operator example")
print(20 in list2)
print("Membership operator example using for loop")
for i in list1:
    print(i)
print("Membership operator in example using for loop")
# To find the number of students got marks greater than or equal to 20.
count=0
for i in list2:
    if i>=20:
        count+=1
print(count)
```

```python
# List comprehension
print("The example for list comprehension.")
cubes=[]
for i in range(1,11):
    cubes.append(i**3)
print("The list which contain the cubes of numbers from 1 to 10 is",cubes)
print("The list comprehension using double list.")
list1=[['Rama',160],['Anya',159],['Mira',158],['sona',161]]
tall=[x for x in list1 if x[1]>=160]
print("The person and theie height whose height is greater or equal to 160",tall)
print("The list comprehension of two lists.")
list1=['a','b','c']
list2=[3,5]
crossproduct = [[x,y] for x in list1 for y in list2]
print("The cross products of the lists are",crossproduct)
```

```python
# Lists as Arguments
'''
Write a code to update an element at given index position.
'''
def listUpdate(a, i, value):
    '''
    Objectivee: to change value at a particular index in list.
    Input Parameters:
        a - list
        i - index of the object in the list to be updated.
        value - modified value at index i.
    Return Value: None
    '''
    a[i]=value
def main():
    '''
    Objective: To change value at a particular index in list
    Input Parameter: None
    Return Value: None
    '''
    list1=[10, 20, 30, [40, 50]]
    listUpdate(list1,1,15)
    print(list1)
if __name__=="__main__":
    main()
```

```python
# Copying list objects
import copy
list1=[10, 20, [30, 40]]
list2=list1
print("The elements of list2 is",list2)
list3=copy.copy(list1)
'''
The copy function creates a shallow copy i.e. it does not create copies of the nested
objects, which doesn't effect the copy list when changes made in original list but when
changes made in nested list of original list the copy list also get effected.
'''
print("The elements of list3 is",list3)
list1[1]=25
print("The element in the list1 and list3 after copy is",list1,list3)
list1[2][0]=35
print("The element in the list1 and list3 after copy is",list1,list3)
'''
The deepcopy function is used to create a copy of list objects such that they can be 
copied to new objects.
'''
list4=copy.deepcopy(list1)
print("The elements in list1 and list4 is",list1,list4)
list1[2][0]=40
print("The elements in list1 and list4 is",list1,list4)
```

```python
# map, reduce and filter operations on a Sequence
from functools import reduce
'''
map:
- map function is used for transforming every value in a given sequence by applying a
function to it.
- Syntax:
    - map(function,iterable object)
'''
# Write a code to insert elements into the list using map function.
a=list(map(int,input("Enter the elements in the list:").split()))
print("The list is",a)
'''
lambda:
- lambda function is used to be applied may have been defined already, or it is defined 
using a lambda expressin which returns a function object.
- Syntax:
    - lambda arguments: expression
'''
# Write a code to find the cube of a number from user as a input using lambda function
cube=lambda x:x**3
print(cube(3))
# Write a code to perform cube on a list using map and lambda function.
lstcubes=list(map(lambda x:x**3,a))
print(lstcubes)
# Write a code to perform sum of elements in the list.
sum1=reduce(lambda x,y:x+y,lstcubes)
print(sum1)
# Write a code to find the even nnumbers i the list using map and filter function.
evencubes=list(filter(lambda x:x%2==0,lstcubes))
print(evencubes)
# Wri a code to find the sum of even cubes using reduce,lambda,filter function.
sumEvenCubes=reduce(lambda x,y:x+y,filter(lambda x:x%2==0,lstcubes))
print(sumEvenCubes)
```

### Sets

```python
# Sets
'''
Sets:
- A set is a mathematics refers to an unordered collection of objects without any 
duplicates and sets is immutable.
- Syntax:
    - var=set()
'''
a=set('aeiou')
print("The example of set is",a)
'''
- A set can contain objects of types such as int, float, str, and tuple but not lists. We 
cannot access elements in a set through indexing or slicing. We cannot apply + and * 
operators on sets.
'''
# Write a code to perform membership operator in sets.
print("The vowels are ", end="")
for v in a:
    print(v,end=" ")
```

```python
# Set functions add, update, remove, pop, and clear.
vehicles={'Scooter', 'Car', 'Bike', 'Bicycle'}
'''
add:
- To add elements to a set we use "add" function.
- Syntax:
    - set.add(obj)
'''
print("The original set is",vehicles,"\n")
vehicles.add('Bus')
print("The updated set is",vehicles,"\n")
'''
update:
- The update function is used to add object such as list, set, range, tuple.
- Syntax:
    - set.update(obj)
'''
print("The original set is",vehicles,"\n")
vehicles.update(['Truck', 'Rickshaw', 'Bike'])
print("The updated set is",vehicles,"\n")
'''
remove:
- The remove function is used to delete a oject from the set.
- Syntax:
    - set.remove(obj)
'''
print("The original set is",vehicles,"\n")
vehicles.remove('Bike')
print("The updated set is",vehicles,"\n")
'''
pop:
- The pop function is used to delete last object in the set.
- Syntax:
    - set.pop()
'''
print("The original set is",vehicles,"\n")
vehicles.pop()
print("The updated set is",vehicles,"\n")
'''
clear:
- The clear function is used to delete all eobjects in the set.
- Syntax:
    - set.clear()
'''
print("The original set is",vehicles,"\n")
vehicles.clear()
print("The set after clear function performed is",vehicles,"\n")
```

```python
# Set function unioon, intersection, difference, and symmetric_difference
fruits={'Apple', 'Orange', 'Tomato', 'Cucumber', 'Watermelon'}
vegetables = {'Cucumber', 'Potato', 'Tomato', 'Watemelon', 'Cauliflower'}
print("The orginal fruits set",fruits,"\n")
print("The original vegetables set",vegetables,"\n")
'''
union:
- The union is used to merge two sets into one set without repeated objects in the sets.
- Syntax:
    - set1.union(set2) or set1|set2
'''
eatables = fruits.union(vegetables)
eatables1=fruits|vegetables
print("The union set of fruits and vegetables is",eatables)
print("The union of set of fruits and vegetables is",eatables1)
print()
'''
intersection:
- The intersection is used to merge two sets into one sets with only common elements.
- Syntax:
    - set1.intersection(set2) or set1 & set2
'''
fruitsAndVegetables = fruits.intersection(vegetables)
print("The intersection set of fruits and vegetables is",fruitsAndVegetables)
fruitsAndVegetables1 = fruits & vegetables
print("The intersection set of fruits and vegetables is", fruitsAndVegetables1)
print()
'''
difference:
- The difference is used to remove common elements and print the remaining elements in the
first set.
- Syntax:
    - set1.difference(set2) or set1 - set2
'''
OnlyFruits = fruits.difference(vegetables)
print("The difference set of fruits and vegetables is",OnlyFruits)
OnlyFruits1 = fruits - vegetables
print("The difference set of fruits and vegetables is",OnlyFruits1)
print()
'''
symmetric_difference:
- The symmetric difference is used to determine the elements that are in one of the either
sets but not in both.
- Syntax:
    - set1.dymmetric_difference(set2)
'''
fruitsXorVegetables = fruits.symmetric_difference(vegetables)
print("The symmetric difference set of fruits and vegetables is",fruitsXorVegetables)
print()
'''
set.union, set.intersection, set.difference is used to perform set operations of union, 
intersection, difference on multiple sets.
'''
digits1 = {0, 1, 2, 3}
digits2 = {2, 4, 5, 6}
digits3 = {0, 7, 8, 9, 2}
print("The union of three sets",set.union(digits1, digits2, digits3))
print()
print("The intersection of three sets", set.intersection(digits1, digits2, digits3))
print()
print("The difference of three sets", set.difference(digits1, digits2, digits3))
'''
copy:
- The copy function can be used to copy set in a similar way to list.
- Syntax:
    - set1.copy()
'''
digits = {0,1,2,3,7,8,9}
digits2 = digits.copy()
print("The set digits is",digits," and copy of digits set is",digits2,"\n")
digits.update({4, 5, 6})
print("The updated digits is",digits," and the copy of digits det is",digits2,"\n")
multiples2 = {2, 4, 6, 8, 10, 12, 14}
multiples4 = {4, 8, 12}
'''
subset:
- The subset is used to check whether the one set is subset of other set.
- Syntax:
    - set.issubset(set1) or set<=set1
'''
print("The multiples2 is subset of multiples4",multiples4 <= multiples2)
print("The multiples2 is subset of multiples4",multiples2.issubset(multiples4))
print()
print("The multiples2 is superset of mulitples4",multiples4 >= multiples2)
print("The multiples2 is superset of multiples4",multiples2.issuperset(multiples4))
```

- Syntax:

```python
month = {}
month[1] = 'Jan'
month[2] = 'Feb'
month[3] = 'Mar'
month[4] = 'Apr'
print("The dictionary month is",month)
print()
print(type(month))
print()
print("The keys in dictionary month is",month.keys())
print()
print("The values in dictionary month is",month.values())
print()
print("The items in dictionary month is",month.items())
```

```python
# Dictionary Operations
import copy
digits = {0:'Zero', 1:'One', 2:'Two', 3:'Three', 4:'Four', 5:'Five', 6:'Six', 7:'Seven',
          8:'Eight', 9:'Nine'}
'''
Length operator len:
- The length operator is used find the the no.of keys-values in dict.
- Syntax:
    - len(dict)
'''
print(f"The length of dict digits is {len(digits)}")
print()
'''
indexing:
- The indexing is used to fetch a element from the tuple based on index.
- Syntax:
    - dict[index]
'''
print(f"The indexing of dict digits is {digits[2]}")
print()
'''
min:
- The min function is used to find the minimun element in the dict.
- Syntax:
    - min(dict)
'''
print(f"The minimum element in dict digits is {min(digits)}")
print()
'''
max:
- The max function is used to find the maximum element in the dict.
- Syntax:
     max(dict)
'''
print(f"The maximum element in dict digits is {max(digits)}")
print()
'''
sum:
- The sum function is used to find sum of elements in the dict.
- Syntax:
    - sum(dict)
'''
print(f"The sum of elements i dict digits is {sum(digits)}")
print()
'''
membership operator in:
- The membership operator is used check whether the element in the dict.
- Syntax:
    - obj in dict
'''
print(f"The element 5 is  in the dict digits {5 in digits}")
print()
'''
del operator:
- The del operator is used to delete a key-value pair from the dictionary.
- Syntax:
    - del dict[key]
'''
print("The key-value corresponding to the key 5 from the dictionary digits del digits[5]")
print()
'''
get:
- The get operator is uesd to extract value corresponding to a given key.
- Syntax:
    - dict(key,parameter)
        - if key is not found in dictionary then parameter is printed else value is printed
        if parameter is not defined then None will be printed as output.
'''
print("The value corresponding to the key 5 in dictionary digits",digits.get(5))
print()
'''
update:
- The update function is used to insert in a dictionary.
- Syntax:
    - dict.update(obj)
'''
moreDigits = {5:'Five', 10:'Ten', 11:'Eleven'}
digits.update(moreDigits)
print("The digits after the updated dictionary digits is",digits)
'''
copy:
- The copy is used to copy dict to other dict.
- Syntax:
    - dict.copy()
'''
newDigits = digits.copy()
print(id(newDigits), id(digits))
```

```python
# Inverted Dictionary
def buildInvDict(dict1):
    '''
    Objective: To construct inverted dictionary.
    Input Parameters: dict1 - dictionary
    Return Value: InvDict - dictionary
    '''
    InvDict = {}
    for key, value in dict1.items():
        if value in InvDict:
            InvDict[value].append(key)
        else:
            InvDict[value] = [key]
    return InvDict
def main():
    '''
    Objective: To find inverted dictionary
    Input Parameters: None
    Return Value: None
    '''
    Numbers = buildInvDict(digits)
    print('Inverted Dictionary:\n', Numbers)
if __name__=="__main__":
    main()
```
