```python
# 4. Write a Python program to find the index of specified item in a list.
a=list(map(int,input("Enter the elements of the list").split()))
b=int(input("Enter the element to find:"))
for i in range(len(a)):
    if b==a[i]:
        print("The element",b,"is found at position",i)
```

```python
# 5.Write a Python program to print the sum of all the items in a list.
a=list(map(int,input("Enter the elements in the list:").split()))
sum=0
for i in a:
    sum+=i
print("The sum of elements in the list is",sum)
```

```python
# 6. Write a Python program to multiply all the items in a list.
a=list(map(int,input("Enter the elements in the list:").split()))
prod=1
for i in a:
    prod*=i
print("The product of elements in the list is",prod)
```

```python
#  7. Write a Python program to get the largest number from a list.
a=list(map(int,input("Enter the elements in the list:").split()))
max=a[0]
for i in a:
    if max<i:
        max=i
print("The maximum element in the list is",max)
```

```python
# 15. Write a Python program to select an item randomly from a list.
import random
a=list(map(int,input("Enter the elements in the list:").split()))
b=random.choice(a)
print("The random element from the list is",b)
```

```python
# 16. Write a Python program to get unique values from a list.
a=list(map(int,input("Enter the elements in the list:").split()))
b=[]
for i in a:
    if i not in b:
        b.append(i)
print("The list of unique numbers:",b)
```
