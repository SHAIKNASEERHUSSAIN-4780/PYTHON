```python
import matplotlib.pyplot as plt
def plotFunctions(a, b, step):
    '''
    Objective: To plot functions f(x) = x**2 and f(x) = x**3 in the same figure
    Input Parameter: a, b, step - numeric value
    Return Value: None
    '''
    nSteps = int((b - a)/step)
    x = [a + step * i for i in range(nSteps+1)]
    y1 = [t**2 for t in x]
    y2 = [t**3 for t in x]
    plt.plot(x, y1, 'ro--', label = 'X vs X**2')
    plt.plot(x, y2, 'ro--', label = 'X vs x**3')
    plt.legend()
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('X vs X**2 and X vs X**3')
    plt.grid()
    plt.show()
def main():
    '''
    Objective: To plot two functions on same graph based on user input
    Input Parameter: None
    Return Value: None
    '''
    a = float(input("Enter first element of the range: "))
    b = float(input("Enter last element of the range: "))
    step = float(input("Enter step size: "))
    plotFunctions(a, b, step)
if __name__ == "__main__":
    main()
```

### Histogram and Pi Chart
- A histogram is used to represent the frequency of various values in a data set using bars, also known as bins.

##### Histogram

```python
import matplotlib.pyplot as plt
def plotHistogram(data):
    '''
    Objective: To plot a histogram
    Input Parameter: data - list
    Return Parameter: None
    '''
    plt.hist(data)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title("Histogram")
    # For setting limits of x axis
    plt.xlim(min(data) - 1, max(data) + 1)
    plt.show()
def main():
    '''
    Objective: To plot a histogram based on user input
    Input Parameter: None
    Return Value: None
    '''
    data = list(map(int, input("Enter the data to be plotted as histogram: ").split()))
    plotHistogram(data)
if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
def plotHist(data):
    '''
    Objective: To plot a histogram with different bins boundaries.
    Input Parameter: data - list
    Return Value: None
    '''
    plt.hist(data, bins = [i - 0.5 for i in range(0, max(data) + 2)])
    plt.xlabel("value")
    plt.ylabel("Frequency")
    plt.title("Histogram")
    plt.xlim(min(data) - 1, max(data) + 1)
def main():
    '''
    Objective: To plot a histogram with different bins boundaries
    Input Parameter: None
    Return Value: None
    '''
    data = list(map(int, input("Enter the data to be plotted for histogram").split()))
    plotHist(data)
if __name__ == "__main__":
    main()
```

##### Pie chart

```python
import matplotlib.pyplot as plt
def plotPieChart(data, labels):
    '''
    Objective: To plot a pie chart
    Input Parameter: data, labels - list
    Return Value: None
    '''
    plt.pie(data, labels = labels, autopct = '%.2f')
    plt.title("Pie Chart")
    plt.show()
def main():
    '''
    Objective: To plot a pie chart
    Input Parameter: None
    Return Value: None
    '''
    data = list(map(int, input("Enter data to be plotted as pie chart: ").split()))
    labels = list(map(str, input("Enter the labels: ").split()))
    plotPieChart(data, labels)
if __name__ == "__main__":
    main()
```

### Sine and Cosine Curves

```python
from matplotlib.pyplot import *
from math import *
def sineCurve():
    '''
    Objective: To plot a sine function
    Input Parameter: None
    Return Value: None
    '''
    subplot(2, 1, 1)
    degrees = range(0, 360 + 1)
    sineValues = [sin(radians(i)) for i in degrees]
    plot(degrees, sineValues)
    xlabel('Degree')
    ylabel('Sine Value')
    title('Sine Curve')
    grid()
    show()
def coseCurve():
    '''
    Objective: To plot a cose function
    Input Parameter: None
    Return Value: None
    '''
    subplot(2, 1, 1)
    degrees = range(0, 360 + 1)
    coseValues = [cos(radians(i)) for i in degrees]
    plot(degrees, coseValues)
    xlabel("Degree")
    ylabel("Cosine Value")
    title("Cosine Value")
    grid()
    show()
def main():
    '''
    Objective: To plot sine and cose curves
    Input Parameter: None
    Return Value: None
    '''
    sineCurve()
    coseCurve()
    tight_layout()
if __name__ == "__main__":
    main()
```

### Graphical Objects: Circle, Ellipse, Rectangle, Polygon, and Arrow

##### Circle

```python
from matplotlib.pyplot import *
from matplotlib.patches import *
def circle(radius):
    '''
    Objective: To plot a circle of given radius
    Input Parameter: radius - numeric
    Return Value: None
    '''
    circle = Circle((0, 0), radius, facecolor='green', edgecolor='red', linestyle='dotted', linewidth=2.2)
    gca().add_patch(circle)
    axis('scaled')
    title('Circle')
    show()
def main():
    '''
    Objective: To plot a circle based on user input
    Input Parameter: None
    REturn Value: None
    '''
    radius = float(input("Enter the radius: "))
    circle(radius)
if __name__ == "__main__":
    main()
```

##### Ellipse

```python
from matplotlib.pyplot import *
from matplotlib.patches import *
def plotEllipse(width, height):
    '''
    Objective: To plot an ellipse of given width and height
    Input Parameter: width, height - numeric
    Return Value: None
    '''
    ellipse = Ellipse((0, 0), width, height, angle = 15, facecolor = 'cyan', edgecolor = 'red', linestyle = 'dashed', linewidth = 2.2)
    gca().add_patch(ellipse)
    axis('scaled')
    title('Ellipse')
    show()
def main():
    '''
    Objective: To plot an ellipse based on user input
    Input Parameter: None
    Return Value: None
    '''
    width = float(input("Enter the width: "))
    height = float(input("Enter the height: "))
    plotEllipse(width, height)
if __name__ == "__main__":
    main()
```

##### Rectangle

```python
from matplotlib.pyplot import *
from matplotlib.patches import *
def plotRectangle(length, breadth):
    '''
    Objective: To plot a rectangle of lenght and breadth
    Input Parameters: lenght and breadth - numeric
    Return Value: None
    '''
    rectangle = Rectangle((0, 0), length, breadth, facecolor = 'green', edgecolor = 'red', linestyle = 'dashed', linewidth = 2.2)
    gca().add_patch(rectangle)
    axis('scaled')
    title('Rectangle')
    show()
def main():
    '''
    Objective: To plot a rectangle based on user input
    Input Parameter: None
    Return Value: None
    '''
    length = float(input("Enter the length: "))
    breadth = float(input("Enter the breadth: "))
    plotRectangle(length, breadth)
if __name__ == "__main__":
    main()
```

##### Polygon

```python
from matplotlib.pyplot import *
from matplotlib.patches import *
def plotPolygon(points):
    '''
    Objective: To plot a polygon joining given a list of points
    Input Parameter: points - list
    Return Value: None
    '''
    polygon = Polygon(points, ec = 'red', linestyle = 'dashdot', lw = 2.2)
    gca().add_patch(polygon)
    axis('scaled')
    title('Polygon')
    show()
def main():
    ''''
    Objective: To plot a polygon based on user input
    Input Parameter: None
    Return Value: None
    '''
    points = eval(input("Enter the points: "))
    plotPolygon(points)
if __name__ == "__main__":
    main()
```

##### Arrows

```python
from matplotlib.pyplot import *
from matplotlib.patches import *
def plotArrow(pos, dx, dy):
    '''
    Objective: To plot an arrow beginning base position pos (x, y) and ending the head position defined by dx and dy relative to the badde of the arrow
    Input Parameters: pos, dx, dy - numeric
    Retrn Value: None
    '''
    x, y = pos[0], pos[1]
    arrow = Arrow(x, y, dx, dy, fill = False, hatch = 'O')
    gca().add_patch(arrow)
    axis('scaled')
    title("Arrow")
    show()
def main():
    '''
    Objective: To plot an arrow based on user input
    Input Parameter: None
    Return Value: None
    '''
    pos = tuple(map(float, input("Enter the starting position (x, y): ").split()))
    dx = float(input("Enter the dx: "))
    dy = float(input("Enter the dy: "))
    plotArrow(pos, dx, dy)
if __name__ == "__main__":
    main()
```

### Animation - Bouncing Ball
