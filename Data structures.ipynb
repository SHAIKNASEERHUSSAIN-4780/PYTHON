```python
# Binary Search
def sort(a):
    for i in range(len(a)):
        for j in range(len(a)):
            if a[i]<a[j]:
                a[i],a[j]=a[j],a[i]
    return a
def binarysearch(a,n):
    low,high=0,len(a)-1
    if n in a:
        while True:
            mid=(low+high)//2
            if a[mid]>n:
                low,high=0,mid-1
            elif a[mid]<n:
                low,high=mid+1,len(a)-1
            else:
                print("The element",n,"is found at the position",mid+1)
                break
    else:
        print("The element is not found")
a=list(map(int,input("Enter the elements into the list:").split()))
sort(a)
print("The sorted array is:",a)
n=int(input("Enter the number to search in the array:"))
binarysearch(a,n)
```

```python
def binarysearch(a,low,high,n):
    if n in a:
        mid=(low+high)//2
        if a[mid]>n:
            binarysearch(a,low,mid+1,n)
        elif a[mid]<n:
            binarysearch(a,mid-1,high,n)
        else:
            print("The element",n,"is found at the position",mid+1,"in the list")
    else:
        print("The element is not found")
a=list(map(int,input("Enter the elements into the list:").split()))
a.sort()
print("The sorted array is:",a)
n=int(input("Enter the element to find in the list:"))
low,high=0,len(a)-1
binarysearch(a,low,high,n)
```

# sorting

```python
# selection sort
def selectionsort(a):
    for i in range(len(a)):
        for j in range(i,len(a)):
            if a[i]>a[j]:
                a[i],a[j]=a[j],a[i]
    return a
a=list(map(int,input("Enter the elements in the list:").split()))
selectionsort(a)
```

```python
# Bubble sort
def bubblesort(a):
    for i in range(len(a)):
        for j in range(0,len(a)-i-1):
            if a[j]>a[j+1]:
                a[j],a[j+1]=a[j+1],a[j]
    return a
a=list(map(int,input("Enter the elements in the list:").split()))
bubblesort(a)
```

```python
# Insertion sort
def insertionsort(a):
    for i in range(len(a)):
        key=a[i]
        j=i-1
        while j>=0 and key<a[j]:
            a[j+1]=a[j]
            j=j-1
        a[j+1]=key
    return a
a=list(map(int,input("Enter the elements into the list:").split()))
insertionsort(a)
```

```python
# mergesort
def mergesort(a):
    if len(a)>1:
        mid=len(a)//2
        b=a[:mid]
        c=a[mid:]
        mergesort(b)
        mergesort(c)
        i=j=k=0
        while i<len(b) and j<len(c):
            if b[i]<c[j]:
                a[k]=b[i]
                k+=1
                i+=1
            elif b[i]>c[j]:
                a[k]=c[j]
                k+=1
                j+=1
        while i<len(b):
            a[k]=b[i]
            i+=1
            k==1
        while j<len(c):
            a[k]=c[j]
            j+=1
            k+=1
    return a
a=list(map(int,input("Enter the elements into the list:").split()))
mergesort(a)
```

## Binarytree

```python
class Node:
    def __init__(s,data):
        s.data=data
        s.left=None
        s.right=None
class Binarytree:
    def __init__(s):
        s.root=None
    def insert(s,data):
        new_node=Node(data)
        if s.root is None:
            s.root=new_node
        else:
            current_node=s.root
            while True:
                if current_node.data>data:
                    if current_node.left is None:
                        current_node.left=new_node
                        break
                    else:
                        current_node=current_node.left
                else:
                    if current_node.right is None:
                        current_node.right=new_node
                        break
                    else:
                        current_node=current_node.right
    def preorder(s,node=None):
        if node is None:
            return
        print("preorder:",node.data)
        s.preorder(node.left)
        s.preorder(node.right)
    def inorder(s,node=None):
        if node is None:
            return
        s.inorder(node.left)
        print("inorder:",node.data)
        s.inorder(node.right)
    def postorder(s,node=None):
        if node is None:
            return
        s.postorder(node.left)
        s.postorder(node.right)
        print("postorder:",node.data)
class Main:
    b = Binarytree()
    b.insert(10)
    b.insert(5)
    b.insert(15)
    b.insert(3)
    b.insert(7)
    b.insert(12)
    b.insert(18)
    
    b.preorder(b.root)
    b.inorder(b.root)
    b.postorder(b.root)
```

## Graph

```python
class Graph:
    def __init__(s,num_vertices):
        s.num_vertices=num_vertices
        s.adj_matrix=[[0]*num_vertices for _ in range(num_vertices)]
    def add_edges(s,i,j):
        if 0<=i<s.num_vertices and 0<=j<s.num_vertices:
            s.adj_matrix[i][j]=1
            s.adj_matrix[j][i]=1
    def print_matrix(s):
        for row in s.adj_matrix:
            print(" ".join(map(str,row)))
g1=Graph(5)
g1.add_edges(0,1)
g1.add_edges(0,3)
g1.add_edges(1,2)
g1.add_edges(1,3)
g1.add_edges(2,4)
g1.add_edges(3,4)
g1.print_matrix()
```
