```python
# 13.Print Pascal's Triangle for N rows using nested loops.
import numpy as np
n=int(input("Enter the no.of rows:"))
a=np.empty((n,n))
for i in range(n):
    for j in range(i+1):
        if(i==0 or j==i):
            a[i][j]=1
        else:
            a[i][j]=a[i-1][j-1]+a[i-1][j]
for i in range(n):
    for j in range(i+1):
        print(int(a[i][j]),end=" ")
    print("\n")
```

```python
# 14.pyramidÂ patterns
n=int(input("Enter the no.of rows:"))
# Pyramid pattern of numbers
for i in range(1,n+1):
    for j in range(1,i+1):
        print(j,end=" ")
    print("")
# Inverted pyramid pattern of numbers
for i in range(0,n):
    for j in range(0,n-i):
        print(n-i,end=" ")
    print()
# Butterfly pattern
for i in range(1,n*2+1):
    for j in range(1,n*2+1):
        if (j<=i and j<=n and j<=2*n-i+1) or (j>n and j>=i and j<=2*n and i+j>=2*n+1):
            print(j,end=" ")
        else:
            print(end="  ")
    print()
```

# OOPS

```python
# Program add two no's using class and object
class add:
    x=100
    y=200
    def sum1(s):
        print("Addition of x and y:",s.x+s.y)
obj=add()
obj.sum1()
```

```python
# Single leve inheritence
class a:
    def display(s):
        print("Display")
class b(a):
    def display1(s):
        print("Show")
c1=b()
c1.display1()
c1.display()
```

```python
# Multi level inheritence
class a:
    def display(s):
        print("Display")
class b(a):
    def display1(s):
        print("Display1")
class c(b):
    def display2(s):
        print("Display2")
class d:
    obj=c()
    obj.display()
    obj.display1()
    obj.display2()
```

```python
# multiple ingeritence
class a:
    def display(s):
        print("Display")
class b:
    def display1(s):
        print("Display1")
class c(a,b):
    def display2(s):
        print("Display2")
class d:
    obj=c()
    obj.display()
    obj.display1()
    obj.display2()
```

```python
# Hierarchical inheritence
class a:
    def display(s):
        print("Display")
class b(a):
    def display1(s):
        print("Display1")
class c(a):
    def display2(s):
        print("Display2")
class d:
    obj=b()
    obj.display()
    obj.display1()
    obj1=c()
    obj1.display()
    obj1.display2()
```

```python
# Hybrid inheritance
class a:
    def display(s):
        print("Display")
class b(a):
    def display1(s):
        print("Display1")
class c(a):
    def display2(s):
        print("Display2")
class d(b,c):
    def display3(s):
        print("Display3")
class e:
    obj=d()
    obj.display3()
    obj.display2()
    obj.display1()
    obj.display()
```

```python
# Cyclic inheritance
class a:
    print("a",end=" ")
    pass
class b(a):
    def dis(s):
        print("b",end=" ")
        super(a,s).dis()
class a(b):
    def dis(s):
        print("a",end=" ")
        super(b,s).dis()
class b(a):
    def dis(s):
        print("a",end=" ")
        super(a,s).dis()
class c:
    obj=b()
    obj.dis()
```

```python
# Abstract class 
from abc import ABC,abstractmethod
class a(ABC):
    @abstractmethod
    def display(s):
        None
class b(a):
    def display(s):
        print("Display")
obj=b()
obj.display()
```

```python
# Interface
from zope import interface as i
class a(i.Interface):
    def square(s,x):
        pass
    def display(s):
        pass
@i.implementer(a)
class b:
    def square(s,x):
        print(x**2)
    def display(s):
        print("Display")
class c:
    obj=b()
    obj.square(5)
    obj.display()
```

```python
# Method overloading
class a:
    def display(s,datatype,*args):
        if datatype=='int':
            s.answer=0
        if datatype=='str':
            s.answer=""
        for i in args:
            s.answer+=i
        print(s.answer)
class b:
    obj=a()
    obj.display('int',1,2)
    obj.display('str','1','2')
```

```python
# Method overriding
class a:
    def display(s):
        print("Display")
class b(a):
    def display(s):
        print("Show")
class c:
    obj=b()
    obj.display()
```
