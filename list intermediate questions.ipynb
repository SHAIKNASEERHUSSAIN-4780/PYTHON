```python
# 1. Write a Python program to count the number of strings where the string length is 2 or more and the first and last characters of a string are same from the given data.
a=list(map(str,input("Enter the elements in the list:").split()))
count=0
for i in a:
    if len(i)>2:
        if i[0]==i[len(i)-1]:
            count+=1
print(count)
```

```python
# 2. Write a Python function to find the list of words which are longer than the specified length from the given data.
s=input("Enter the string:")
n=int(input("Enter the legnth of string we need:"))
s=s.split()
for i in s:
    if len(i)<=n:
        s.remove(i)
print(s)
```

```python
# 3. Write a Python function that takes two parameters as lists and returns True if they have at least one common element.
def common(a,b):
    for i in a:
        if i in b:
            return True
    return None
a=list(map(int,input("Enter the elements in the list:").split()))
b=list(map(int,input("Enter the elements in the list:").split()))
c=common(a,b)
print("There are common element in the lists is",c)
```

```python
# 4. Write a Python program to generate a 3*4*6 3D array whose each element is '*'. 
import numpy as np
m,n,o=map(int,input("Enter the row,column & height of the 3d array").split())
a=np.empty((m,n,o),dtype=object)
for i in range(m):
    for j in range(n):
        for k in range(o):
            a[i][j][k]='*'
print(a)
```

```python
# 5. Write a Python program to generate and print a list of first and last 5 elements where the values are square of numbers between 1 and 30 (both included).
m,n=map(int,input("Enter the range:").split())
a=[]
b=[]
for i in range(m,n+1):
    if(m-i<=0 and m-i>-5):
        a.append(i**2)
    elif(n-i<5 and n-i>=0):
        b.append(i**2)
print("Squares of first 5 numbers between",m,"to",n,":",a)
print("Squares of last 5 numbers between",m,"to",n,":",b)
```

```python
# 7. Write a Python program to find the second largest number in a list.
from sys import maxsize as ms
def se(a):
    max1=(-ms-1)
    for i in a:
        if max1<i:
            max1=i
    return max1
a=list(map(int,input("Enter the elements in the list:").split()))
b=se(a)
a.remove(b)
b=se(a)
print("The second largest element in the list is",b);
```
