# NumPy Methods Categorized with Descriptions

---

## 1. Array Creation

| `np.ones()`                | Creates an array filled with ones                 |
| `np.full()`                | Creates an array filled with a specified value    |
| `np.eye()`                 | Creates an identity matrix                        |
| `np.arange()`              | Returns evenly spaced values within a range       |
| `np.linspace()`            | Returns evenly spaced numbers over an interval    |
| `np.random.rand()`         | Random values in a given shape (uniform [0,1))    |
| `np.random.randn()`        | Random values from standard normal distribution   |
| `np.random.randint()`      | Random integers within a range                    |

| `np.vstack()`              | Stacks arrays vertically                        |

| `np.multiply()`            | Element-wise multiplication                     |
| `np.divide()`              | Element-wise division                           |
| `np.power()`               | Element-wise exponentiation                     |

| `np.sqrt()`                | Square root of each element                     |

| `np.median()`              | Median of elements                              |
| `np.std()`                 | Standard deviation                              |
| `np.sum()`                 | Sum of array elements                           |

## 4. Logical and Comparison

| Method                     | Description                                     |
|----------------------------|-------------------------------------------------|

---

## 5. Utility Methods

| Method                     | Description                                     |
|----------------------------|-------------------------------------------------|
| `np.unique()`              | Returns unique elements                        |
| `np.sort()`                | Returns sorted array                           |
| `np.argsort()`             | Indices that would sort the array              |
| `np.clip()`                | Clamps values within a range                   |
| `np.isfinite()`            | Checks for finite values                       |
| `np.isnan()`               | Checks for NaN values                          |


```python
import numpy as np

print("üîπ 1. ARRAY CREATION METHODS")

a = np.array([[1, 2], [3, 4]])
print("np.array():\n", a)

zeros = np.zeros((2, 2))
print("np.zeros():\n", zeros)

ones = np.ones((2, 3))
print("np.ones():\n", ones)

full = np.full((2, 2), 7)
print("np.full():\n", full)

identity = np.eye(3)
print("np.eye():\n", identity)

range_arr = np.arange(0, 10, 2)
print("np.arange():", range_arr)

linspace_arr = np.linspace(0, 1, 5)
print("np.linspace():", linspace_arr)

rand_arr = np.random.rand(2, 2)
print("np.random.rand():\n", rand_arr)

randn_arr = np.random.randn(2, 2)
print("np.random.randn():\n", randn_arr)

randint_arr = np.random.randint(1, 10, (2, 2))
print("np.random.randint():\n", randint_arr)

print("\nüîπ 2. ARRAY MANIPULATION")

reshaped = a.reshape((4, 1))
print("np.reshape():\n", reshaped)

flattened = a.flatten()
print("np.flatten():", flattened)

raveled = a.ravel()
print("np.ravel():", raveled)

transposed = a.T
print("np.transpose() / .T:\n", transposed)

vstacked = np.vstack((a, a))
print("np.vstack():\n", vstacked)

hstacked = np.hstack((a, a))
print("np.hstack():\n", hstacked)

concatenated = np.concatenate((a, a), axis=0)
print("np.concatenate():\n", concatenated)

split_arr = np.array_split(range_arr, 2)
print("np.array_split():", split_arr)

print("\nüîπ 3. MATHEMATICAL OPERATIONS")

added = np.add(a, 1)
print("np.add():\n", added)

subtracted = np.subtract(a, 1)
print("np.subtract():\n", subtracted)

multiplied = np.multiply(a, 2)
print("np.multiply():\n", multiplied)

divided = np.divide(a, 2)
print("np.divide():\n", divided)

powered = np.power(a, 2)
print("np.power():\n", powered)

squared_root = np.sqrt(a)
print("np.sqrt():\n", squared_root)

sine = np.sin(a)
print("np.sin():\n", sine)

cosine = np.cos(a)
print("np.cos():\n", cosine)

dot_product = np.dot(a, a)
print("np.dot():\n", dot_product)

mean_val = np.mean(a)
print("np.mean():", mean_val)

median_val = np.median(a)
print("np.median():", median_val)

std_dev = np.std(a)
print("np.std():", std_dev)

total_sum = np.sum(a)
print("np.sum():", total_sum)

max_val = np.max(a)
print("np.max():", max_val)

min_val = np.min(a)
print("np.min():", min_val)

argmax_idx = np.argmax(a)
print("np.argmax():", argmax_idx)

argmin_idx = np.argmin(a)
print("np.argmin():", argmin_idx)

print("\nüîπ 4. LOGICAL AND COMPARISON")

eq = np.equal(a, 2)
print("np.equal():\n", eq)

greater = np.greater(a, 2)
print("np.greater():\n", greater)

less = np.less(a, 2)
print("np.less():\n", less)

where_ex = np.where(a > 2, a, 0)
print("np.where():\n", where_ex)

any_true = np.any(a > 3)
print("np.any():", any_true)

all_true = np.all(a < 5)
print("np.all():", all_true)

print("\nüîπ 5. UTILITY METHODS")

unique_vals = np.unique([1, 2, 2, 3, 3])
print("np.unique():", unique_vals)

sorted_arr = np.sort([3, 1, 2])
print("np.sort():", sorted_arr)

argsorted = np.argsort([3, 1, 2])
print("np.argsort():", argsorted)

clipped = np.clip([1, 3, 5], 2, 4)
print("np.clip():", clipped)

finite_check = np.isfinite([1, np.inf])
print("np.isfinite():", finite_check)

nan_check = np.isnan([np.nan, 1])
print("np.isnan():", nan_check)
```

## üßπ Data Cleaning Methods

Pandas provides powerful tools for handling missing data, duplicates, and data-type issues.

| Method                         | Description                                      |
|--------------------------------|--------------------------------------------------|
| `df.isnull()`                  | Check for missing values                        |
| `df.notnull()`                 | Check for non-missing values                    |
| `df.dropna()`                  | Drop rows with any missing values               |
| `df.fillna(value)`             | Fill missing values with given `value`          |
| `df.replace(old, new)`        | Replace specified values                        |
| `df.duplicated()`              | Returns Boolean Series of duplicate rows        |
| `df.drop_duplicates()`         | Remove duplicate rows                           |
| `df.astype(dtype)`             | Change column data types                        |

Use these tools to ensure data consistency and quality before analysis or modeling.


```python
import pandas as pd
import numpy as np

# Sample CSV-style data using a dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Bob', None],
    'Age': [25, 30, np.nan, 30, 22],
    'City': ['NY', 'LA', 'Chicago', 'LA', 'Boston']
}

# Create DataFrame
df = pd.DataFrame(data)
print("üîπ Original DataFrame:\n", df)

# Write to CSV (simulated with buffer)
from io import StringIO
csv_data = df.to_csv(index=False)
print("\nüîπ CSV Data (written):\n", csv_data)

# Read from CSV (simulate)
df_read = pd.read_csv(StringIO(csv_data))
print("üîπ Read CSV into DataFrame:\n", df_read)

# ----------------- Data Cleaning and Manipulation Methods ------------------

print("\nüîπ Null Check:\n", df.isnull())
filled_df = df.fillna({'Name': 'Unknown', 'Age': df['Age'].mean()})
print("\nüîπ Fill Missing Values:\n", filled_df)
dropped_na = df.dropna()
print("\nüîπ Drop Rows with NA:\n", dropped_na)
replaced_df = df.replace({'City': {'NY': 'New York'}})
print("\nüîπ Replace Values in Column:\n", replaced_df)
print("\nüîπ Check Duplicates:\n", df.duplicated())
no_duplicates = df.drop_duplicates()
print("\nüîπ Drop Duplicates:\n", no_duplicates)
df['Age'] = df['Age'].fillna(0).astype(int)
print("\nüîπ Converted 'Age' to Integer:\n", df)

# ----------------- More DataFrame Operations ------------------

# Rename columns
renamed_df = df.rename(columns={'Name': 'FullName', 'City': 'Location'})
print("\nüîπ Renamed Columns:\n", renamed_df)

# Set/Reset Index
df_indexed = df.set_index('Name')
print("\nüîπ Set 'Name' as Index:\n", df_indexed)
reset_df = df_indexed.reset_index()
print("\nüîπ Reset Index:\n", reset_df)

# Sorting
sorted_by_age = df.sort_values(by='Age', ascending=False)
print("\nüîπ Sorted by Age Descending:\n", sorted_by_age)

# Filtering
over_25 = df[df['Age'] > 25]
print("\nüîπ Rows where Age > 25:\n", over_25)

# Apply custom function
df['AgePlus5'] = df['Age'].apply(lambda x: x + 5)
print("\nüîπ Apply Function (Age + 5):\n", df)

# Mapping
city_map = {'NY': 'New York', 'LA': 'Los Angeles', 'Chicago': 'Chicago', 'Boston': 'Boston'}
df['City'] = df['City'].map(city_map)
print("\nüîπ Mapped City Names:\n", df)

# Drop column
dropped_col = df.drop(columns=['AgePlus5'])
print("\nüîπ Dropped 'AgePlus5' Column:\n", dropped_col)

# Value counts
print("\nüîπ City Value Counts:\n", df['City'].value_counts())

# Group by
grouped = df.groupby('City')['Age'].mean()
print("\nüîπ Grouped by City (Mean Age):\n", grouped)

# Unique values
print("\nüîπ Unique Names:\n", df['Name'].unique())

# Describe statistics
print("\nüîπ Age Statistics:\n", df['Age'].describe())

# ----------------- Utility & Info Methods ------------------

# Data types
print("\nüîπ Data Types:\n", df.dtypes)

# Memory usage
print("\nüîπ Memory Usage (bytes):\n", df.memory_usage(deep=True))

# General info
print("\nüîπ DataFrame Info:")
df.info()

# Shape and size
print("\nüîπ Shape (Rows, Columns):", df.shape)
print("üîπ Total Elements (size):", df.size)

# Head and tail
print("\nüîπ First 3 Rows:\n", df.head(3))
print("\nüîπ Last 2 Rows:\n", df.tail(2))

# Sample rows
print("\nüîπ Random Sample (n=2):\n", df.sample(n=2))

# Column/row access
print("\nüîπ Access 'City' Column:\n", df['City'])
print("\nüîπ Access First Row (iloc):\n", df.iloc[0])
print("\nüîπ Access by Label (loc):\n", df.loc[0])

# Check if column exists
print("\nüîπ 'Age' in columns:", 'Age' in df.columns)

# Column-wise null count
print("\nüîπ Null Count per Column:\n", df.isnull().sum())

# Correlation matrix
print("\nüîπ Correlation Matrix:\n", df.corr(numeric_only=True))

# Drop row by index
dropped_row = df.drop(index=0)
print("\nüîπ Dropped First Row:\n", dropped_row)

# Insert new column
df.insert(1, 'Gender', ['F', 'M', 'M', 'M', 'F'])
print("\nüîπ Inserted 'Gender' Column:\n", df)

# Rename index
df.index = ['row1', 'row2', 'row3', 'row4', 'row5']
print("\nüîπ Renamed Index:\n", df)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Segoe UI Emoji'
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]
random_data = np.random.normal(0, 1, 1000)

# Create main figure
fig, axs = plt.subplots(4, 3, figsize=(18, 16), constrained_layout=True)
fig.suptitle("üîß All-in-One Matplotlib Plotting Showcase", fontsize=18)

# 1. Line Plot with Custom Style & Annotation
axs[0, 0].plot(x, y1, label='sin(x)', color='blue', linestyle='--', linewidth=2, marker='o')
axs[0, 0].set_title('Line Plot with Marker & Annotation')
axs[0, 0].set_xlabel('X Axis')
axs[0, 0].set_ylabel('Y Axis')
axs[0, 0].grid(True)
axs[0, 0].legend()
axs[0, 0].annotate('Peak', xy=(1.5, 1), xytext=(2, 1.5),
                   arrowprops=dict(facecolor='black', shrink=0.05))

# 2. Bar Plot with Custom Colors & Axis Limits
axs[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'orange'])
axs[0, 1].set_title('Bar Plot with Axis Limits')
axs[0, 1].set_ylim(0, 100)

# 3. Horizontal Bar with Text & Rotated Ticks
axs[0, 2].barh(categories, values, color='lightgreen')
axs[0, 2].set_title('Horizontal Bar')
axs[0, 2].set_xticks([0, 25, 50, 75, 100])
axs[0, 2].set_xlabel('Score')
axs[0, 2].text(10, 'B', '‚Üê Label B', color='red')

# 4. Scatter Plot with Custom Markers
axs[1, 0].scatter(x, y2, color='purple', label='cos(x)', marker='*')
axs[1, 0].set_title('Scatter with Custom Marker')
axs[1, 0].legend()

# 5. Histogram with Density
axs[1, 1].hist(random_data, bins=20, color='pink', density=True, edgecolor='black')
axs[1, 1].set_title('Histogram with Density')

# 6. Box Plot with Notch
axs[1, 2].boxplot([random_data, np.random.normal(1, 2, 1000)], notch=True, patch_artist=True,
                  boxprops=dict(facecolor='lightblue'))
axs[1, 2].set_title('Box Plot (Notched)')
axs[1, 2].set_xticklabels(['Dist1', 'Dist2'])

# 7. Pie Chart with Shadow
axs[2, 0].pie(values, labels=categories, autopct='%1.1f%%', shadow=True, startangle=140)
axs[2, 0].set_title('Pie Chart with Shadow')

# 8. Stack Plot
axs[2, 1].stackplot(x, np.row_stack([np.sin(x), np.sin(x + 1), np.sin(x + 2)]),
                    labels=['A', 'B', 'C'], alpha=0.6)
axs[2, 1].set_title('Stack Plot')
axs[2, 1].legend(loc='upper left')

# 9. Shared Axis Plot
x2 = np.arange(0, 4)
y2a = [1, 4, 9, 16]
y2b = [2, 3, 5, 7]
axs[2, 2].plot(x2, y2a, label='Line A')
axs[2, 2].plot(x2, y2b, label='Line B')
axs[2, 2].set_title('Shared Axes Plot')
axs[2, 2].legend()

# 10. Twin Y-Axis
ax_main = axs[3, 0]
ax_twin = ax_main.twinx()
ax_main.plot(x, y1, 'g-', label='sin(x)')
ax_twin.plot(x, y2, 'b-', label='cos(x)')
ax_main.set_title('Twin Y Axes')
ax_main.set_ylabel('sin(x)', color='g')
ax_twin.set_ylabel('cos(x)', color='b')

# 11. Customize Spines & Ticks
axs[3, 1].plot(x, y1, color='black')
axs[3, 1].set_title('Custom Spines')
axs[3, 1].spines['top'].set_visible(False)
axs[3, 1].spines['right'].set_visible(False)
axs[3, 1].tick_params(axis='x', rotation=45)

# 12. Object-Oriented API with Manual Layout
fig2, ax = plt.subplots(figsize=(5, 4))
ax.plot(x, y2, label='cos(x)', color='brown')
ax.set_title('OOP API Plot')
ax.set_xlabel('X Axis')
ax.set_ylabel('Cosine')
ax.legend()
ax.grid(True)

# Save final output
fig.savefig("full_matplotlib_methods.png")
fig2.savefig("oop_api_plot.png")

plt.show()

# Display data utility
print("üéØ x.shape =", x.shape)
print("üß™ First 5 y1 =", y1[:5])
```

# üìò Seaborn Functions ‚Äì Full Explanation and Categorization

---

## üé® 1. Relational Plots

| **Function**                       | **Description** |
|-----------------------------------|-----------------|
| `scatterplot(data, x, y)`          | Scatter plot for visualizing relationship between two variables. |
| `lineplot(data, x, y)`             | Line plot showing trends over a continuous variable. |
| `relplot(data, x, y, kind="scatter"/"line")` | General relational plot; can produce scatter or line plots with facet support. |

---

## üß∞ 2. Distribution Plots

| **Function**                       | **Description** |
|-----------------------------------|-----------------|
| `histplot(data, x, bins)`          | Histogram to visualize frequency distribution of a variable. |
| `kdeplot(data, x, shade=True)`     | Kernel Density Estimate plot for probability density estimation. |
| `ecdfplot(data, x)`                | Empirical Cumulative Distribution Function plot. |
| `displot(data, x, kind="hist"/"kde")` | General distribution plot with facet options. |

---

## üìä 3. Categorical Plots

| **Function**                       | **Description** |
|-----------------------------------|-----------------|
| `boxplot(data, x, y, hue=None)`    | Box-and-whisker plot showing distribution summary for categories. |
| `violinplot(data, x, y, hue=None, split=False)` | Combines boxplot and KDE for category-wise distribution. |
| `stripplot(data, x, y, hue=None, dodge=False, jitter=True)` | Scatter plot for categorical data points. |
| `swarmplot(data, x, y, hue=None, dodge=False)` | Categorical scatter plot avoiding overlapping points. |
| `barplot(data, x, y, hue=None, ci=95)` | Bar plot with height representing estimate of central tendency and error bars. |
| `countplot(data, x, hue=None)`     | Shows count of observations in each category. |
| `pointplot(data, x, y, hue=None)`  | Shows point estimates and confidence intervals for categorical data. |
| `catplot(data, x, y, kind="box"/"violin"/"strip"/"swarm"/"bar"/"point")` | General categorical plot with faceting options. |

---

## üìè 4. Matrix Plots

| **Function**                       | **Description** |
|-----------------------------------|-----------------|
| `heatmap(data, annot=False, cmap=None)` | Annotated color-coded matrix to visualize correlations or values. |
| `clustermap(data, method="average", metric="euclidean")` | Hierarchically-clustered heatmap. |

---

## üß™ 5. Regression Plots

| **Function**                       | **Description** |
|-----------------------------------|-----------------|
| `regplot(data, x, y)`              | Scatter plot with fitted linear regression line. |
| `lmplot(data, x, y, hue=None)`     | High-level regression plot with optional faceting by row/col. |
| `residplot(data, x, y)`            | Plots residuals to assess regression model fit. |

---

## üåê 6. Multi-Plot Grids

| **Function**                       | **Description** |
|-----------------------------------|-----------------|
| `pairplot(data, hue=None)`          | Pairwise scatter plots for all numerical variables. |
| `pairgrid(data, hue=None)`          | Low-level grid object for creating pairwise plots with flexible customization. |
| `FacetGrid(data, row=None, col=None, hue=None)` | Grid for plotting data subsets by row/column/facet variables. |
| `displot(data, x, col=None, row=None, hue=None)` | Distribution plot with facet support. |
| `catplot(data, x, y, col=None, row=None, hue=None)` | Categorical plot with facet support. |

---

## üé® 7. Style, Themes, and Palettes

| **Function**                       | **Description** |
|-----------------------------------|-----------------|
| `set_style("whitegrid"/"dark"/"ticks")` | Sets background and grid style of plots. |
| `set_context("paper"/"talk"/"notebook")` | Sets scaling of elements for figure context. |
| `set_palette("Set1"/"deep"/"pastel")` | Sets color palette for plots. |
| `color_palette()`                   | Returns current color palette or creates a new one. |
| `despine()`                         | Removes top and right axes spines for cleaner look. |

---

## üíæ 8. Display and Save

| **Function**                       | **Description** |
|-----------------------------------|-----------------|
| `plt.show()`                        | Displays the current plot (from Matplotlib). |
| `savefig("filename.png")`           | Saves the current figure to file (via Matplotlib). |

---

## ‚öôÔ∏è 9. Object-Oriented Style (Seaborn with Matplotlib OOP)

| **Method**                         | **Description** |
|-----------------------------------|-----------------|
| `g = FacetGrid(...); g.map(func, x)` | Map a plotting function to a Seaborn grid object. |
| `g.set_titles()`                    | Sets titles of each subplot in FacetGrid. |
| `g.set_axis_labels(x, y)`           | Sets axis labels for FacetGrid. |
| `g.add_legend()`                     | Adds legend to grid plots. |

## üßæ 10. Utility and Data Inspection

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# =======================
# 1. Relational Plots
# =======================
fig, axes = plt.subplots(1, 3, figsize=(18, 5))  # 3 plots in a row

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="time", size="size", ax=axes[0])
axes[0].set_title("Scatter Plot")

sns.lineplot(data=tips, x="size", y="tip", hue="sex", style="sex", markers=True, dashes=False, ax=axes[1])
axes[1].set_title("Line Plot")

# High-level relplot (separate)
sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter", hue="time")
plt.title("Relplot Scatter")
plt.show()

# =======================
# 2. Distribution Plots
# =======================
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.histplot(tips["total_bill"], bins=20, kde=True, ax=axes[0])
axes[0].set_title("Histogram with KDE")

sns.kdeplot(tips["total_bill"], fill=True, ax=axes[1])
axes[1].set_title("KDE Plot")

sns.ecdfplot(data=tips, x="total_bill", hue="sex", ax=axes[2])
axes[2].set_title("ECDF Plot")

plt.tight_layout()
plt.show()

# High-level displot (separate)
sns.displot(data=tips, x="total_bill", kind="hist")
plt.title("Displot Histogram")
plt.show()

# =======================
# 3. Categorical Plots
# =======================
fig, axes = plt.subplots(2, 3, figsize=(18, 10))  # 2 rows, 3 columns
axes = axes.flatten()

sns.boxplot(data=tips, x="day", y="total_bill", hue="sex", ax=axes[0])
axes[0].set_title("Box Plot")

sns.violinplot(data=tips, x="day", y="tip", hue="sex", split=True, ax=axes[1])
axes[1].set_title("Violin Plot")

sns.stripplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True, jitter=True, ax=axes[2])
axes[2].set_title("Strip Plot")

sns.swarmplot(data=tips, x="day", y="tip", hue="sex", dodge=True, ax=axes[3])
axes[3].set_title("Swarm Plot")

sns.barplot(data=tips, x="day", y="total_bill", hue="sex", errorbar="sd", ax=axes[4])
axes[4].set_title("Bar Plot")

sns.countplot(data=tips, x="day", hue="sex", ax=axes[5])
axes[5].set_title("Count Plot")

plt.tight_layout()
plt.show()

# Point plot and catplot (separate)
sns.pointplot(data=tips, x="day", y="tip", hue="sex")
plt.title("Point Plot")
plt.show()

sns.catplot(data=tips, x="day", y="total_bill", kind="box", hue="sex")
plt.title("Catplot Box")
plt.show()

# =======================
# 4. Matrix Plots
# =======================
numeric_tips = tips.select_dtypes(include='number')
corr = numeric_tips.corr()

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.heatmap(corr, annot=True, cmap="coolwarm", ax=axes[0])
axes[0].set_title("Heatmap of Numeric Features")

sns.clustermap(corr, annot=True, cmap="viridis")  # separate, cannot combine in axes
plt.show()

# =======================
# 5. Regression Plots
# =======================
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.regplot(data=tips, x="total_bill", y="tip", ax=axes[0])
axes[0].set_title("Regplot")

sns.residplot(data=tips, x="total_bill", y="tip", ax=axes[1])
axes[1].set_title("Residual Plot")

# lmplot (high-level)
sns.lmplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.title("Lmplot")
plt.show()

# =======================
# 6. Multi-Plot Grids
# =======================
sns.pairplot(tips, hue="sex")  # separate
plt.show()

g = sns.FacetGrid(tips, col="time", row="sex")
g.map(sns.histplot, "total_bill", bins=15)
plt.show()

# =======================
# 7. Style & Themes
# =======================
sns.set_style("whitegrid")
sns.set_context("talk")
sns.set_palette("Set2")

plt.figure(figsize=(6, 5))
sns.histplot(tips["tip"], kde=True)
plt.title("Styled Histogram")
plt.show()

# =======================
# 8. Utility / Data Inspection
# =======================
print("Head of dataset:\n", tips.head())
print("\nDataset info:")
print(tips.info())
print("\nShape:", tips.shape)
print("\nColumns:", tips.columns)
print("\nDescription:\n", tips.describe())
```
